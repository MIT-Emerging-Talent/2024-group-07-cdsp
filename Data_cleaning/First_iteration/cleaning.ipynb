Dropping Columns that contains irrelevant infromation.

```python
data.drop([col for col in data.columns if col in ['origin_PDI',
 'origin_IDV',
 'origin_MAS',
 'origin_UAI',
 'origin_LTO',
 'origin_area',
  'destination_area',
 'origin_cont_code',
 'destination_cont_code',
'origin_coordinate',
 'origin_latitude',
 'origin_longitude','destination_coordinate',
 'destination_latitude',
 'destination_longitude', 'destination_PDI',
 'destination_IDV',
 'destination_MAS',
 'destination_UAI',
 'destination_LTO',
'sci_2021',
 'sci_2020''destination_fb_users',
 'destination_fb_users_perc']
 ],axis=1,inplace=True)
```

Helper function

```python
#Function to help extract 5 years of data for each of the factors
def dataExtractor(exp):
    result = []
    for i in range(2014,2020):
        newexp = exp+f"{i}"
        
        if newexp in data.columns.to_list():
            result.append(newexp)
    return result
```

Sorting and skimming ESTAT Data set

```python
ESTAT_ordered=["from_to",'origin_country']
for i in range(2014,2020):
    for j in ['T','M','F']:
        for k in ['T','<15','15-64','>64']:
            for l in ['cit' , 'res']:
                exp = f'ESTAT_{i}_{j}_{k}_{l}'
                if exp in data.columns.to_list():
                    ESTAT_ordered.append(exp)

ESTAT_origin_pop = dataExtractor('ESTAT_origin_pop_')
ESTAT_origin_pop_density = dataExtractor('ESTAT_origin_pop_dens_')
ESTAT_destination_pop = dataExtractor('ESTAT_destination_pop_')
ESTAT_destination_pop_density = dataExtractor('ESTAT_destination_pop_dens_')
ESTAT_origin_total_imm_res = dataExtractor('ESTAT_origin_total_imm_res_')
ESTAT_origin_total_em_res = dataExtractor('ESTAT_origin_total_em_res_')
ESTAT_destination_total_imm_res = dataExtractor('ESTAT_destination_total_imm_res_')
ESTAT_destination_total_em_res = dataExtractor('ESTAT_destination_total_em_res_')

```

Sorting and skimming UN Data set

```python
UN_ordered = []
for i in range(2015,2021):
    for j in ['T','M','F']:
        for k in ['T']:
            for l in ['cit' , 'res']:
                exp = f'UN_{i}_{j}_{k}_{l}'
                if exp in data.columns.to_list():
                    UN_ordered.append(exp)
    

UN_origin_pop = dataExtractor('UN_origin_pop_')
UN_destination_pop = dataExtractor('UN_destination_pop_')

UN_origin_total_imm_cit = dataExtractor('UN_origin_total_imm_cit_')

UN_origin_total_imm_res = dataExtractor('UN_origin_total_imm_res_')

UN_origin_total_em_cit = dataExtractor('UN_origin_total_em_cit_')

UN_destination_total_imm_cit = dataExtractor('UN_destination_total_imm_cit_')

UN_destination_total_imm_res = dataExtractor('UN_destination_total_imm_res_')

UN_destination_total_em_cit = dataExtractor('UN_destination_total_em_cit_')
```

Emerging all the lists into one big list

```python
origin_gdp = dataExtractor('origin_gdp_')

destination_gdp = dataExtractor('destination_gdp_')

ordered_columns = [ESTAT_ordered,['origin_name','origin_official_name','origin_cont_name','origin_neighbors','destination_gdp_2020','origin_religion',
 'origin_gdp_2020','origin_languages','destination_name','destination_official_name','destination_cont_name','destination_neighbors','destination_religion',
 'destination_languages','from_to_cont','geodesic_distance_km'], ESTAT_origin_pop, ESTAT_origin_pop_density, ESTAT_destination_pop, ESTAT_destination_pop_density,
  ESTAT_origin_total_imm_res, ESTAT_origin_total_em_res, ESTAT_destination_total_imm_res, ESTAT_destination_total_em_res,
   UN_ordered, UN_origin_pop, UN_destination_pop, UN_origin_total_imm_cit , UN_origin_total_imm_res, UN_origin_total_em_cit,
    UN_destination_total_imm_cit, UN_destination_total_imm_res, UN_destination_total_em_cit, ['UN_origin_net_migr_rate_2000-2005',
 'UN_origin_net_migr_rate_2005-2010','UN_origin_net_migr_rate_2010-2015','UN_origin_net_migr_rate_2015-2020','UN_origin_net_migr_rate_2020-2025',
 'UN_origin_net_migr_2000-2005','UN_origin_net_migr_2005-2010','UN_origin_net_migr_2010-2015','UN_origin_net_migr_2015-2020','UN_origin_net_migr_2020-2025',
 'UN_destination_net_migr_rate_2000-2005','UN_destination_net_migr_rate_2005-2010','UN_destination_net_migr_rate_2010-2015','UN_destination_net_migr_rate_2015-2020',
 'UN_destination_net_migr_rate_2020-2025','UN_destination_net_migr_2000-2005','UN_destination_net_migr_2005-2010','UN_destination_net_migr_2010-2015',
 'UN_destination_net_migr_2015-2020','UN_destination_net_migr_2020-2025','UN_migr_stocks_2000_T_T','UN_migr_stocks_2005_T_T','UN_migr_stocks_2010_T_T','UN_migr_stocks_2015_T_T','UN_migr_stocks_2020_T_T',
 'UN_migr_stocks_2000_M_T','UN_migr_stocks_2005_M_T','UN_migr_stocks_2010_M_T','UN_migr_stocks_2015_M_T','UN_migr_stocks_2020_M_T','UN_migr_stocks_2000_F_T',
 'UN_migr_stocks_2005_F_T','UN_migr_stocks_2010_F_T','UN_migr_stocks_2015_F_T','UN_migr_stocks_2020_F_T','origin_alpha_3',
 'destination_alpha_3'], origin_gdp, destination_gdp ]


#flattening the nested list
ordered_columns = sum(ordered_columns,[])
```

Creating a new dataset file with rearranged columnms

```python
ordered_Data =  data[ordered_columns]
ordered_Data.to_csv("ordered_mimi.csv")
```
