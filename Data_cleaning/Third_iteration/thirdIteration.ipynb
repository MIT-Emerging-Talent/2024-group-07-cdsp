```python
import pandas as pd    
from pandas.api.types import is_string_dtype
from pandas.api.types import is_numeric_dtype
```

```python
population = pd.read_csv("world_population.csv")
main_dataset = pd.read_csv("merged_Dataset.csv")
```

```python
population_cols = population[["CCA3", "2015 Population"]]
main_dataset_alphas= main_dataset['origin_alpha_3'].unique().tolist()
```

```python
popul_ation= population['CCA3'].unique().tolist()

missing_origin = [i for i in main_dataset_alphas if i not in popul_ation]
```

```python
origin_pop = population_cols[population_cols['CCA3'].isin(main_dataset_alphas)]
origin_pop_frame = pd.DataFrame(origin_pop)
origin_pop_frame.reset_index(drop=True, inplace=True)
origin_pop_frame = origin_pop_frame.rename(columns={'CCA3':'origin_alpha_3', '2015 Population':'origin_pop_2015'})
Pop_merged = main_dataset.merge(origin_pop_frame,on = 'origin_alpha_3', how = 'left')
Pop_merged

```

```python
Pop_merged.to_csv("Merged_Dataset.csv", index=False)
```
