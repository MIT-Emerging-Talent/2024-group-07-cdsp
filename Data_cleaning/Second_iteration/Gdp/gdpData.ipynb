```python
import pandas as pd    
from pandas.api.types import is_string_dtype
from pandas.api.types import is_numeric_dtype
```

```python
gdp_countries = pd.read_excel("Old/original_P_Data_Extract_From_World_Development_Indicators.xlsx")
countries_of_interest = pd.read_csv("countries_of_interest_data.csv", low_memory=False)
gdp_columns = ['Country Code', '2014 [YR2014]', '2015 [YR2015]', '2016 [YR2016]','2017 [YR2017]','2018 [YR2018]','2019 [YR2019]'] 

gdp_c= gdp_countries['Country Code'].unique().tolist()
```

Creating a seperate Dataframe for Origin GDP from the GDP Data set in the "Old" Folder

```python
countries_origin = countries_of_interest['origin_alpha_3'].unique().tolist()

missing_origin = [i for i in countries_origin if i not in gdp_c]


origin_gpd = gdp_countries[gdp_countries['Country Code'].isin(countries_origin)]

origin_gpd = origin_gpd[gdp_columns]
origin_gpd_frame = pd.DataFrame(origin_gpd)
origin_gpd_frame.reset_index(drop=True, inplace=True)
origin_gpd_frame = origin_gpd_frame.rename(columns={'Country Code':'origin_alpha_3', '2014 [YR2014]':'origin_gdp_2014', '2015 [YR2015]':'origin_gdp_2015', '2016 [YR2016]':'origin_gdp_2016','2017 [YR2017]':'origin_gdp_2017','2018 [YR2018]':'origin_gdp_2018','2019 [YR2019]':'origin_gdp_2019'})
origin_gpd_frame


```

Creating a seperate Dataframe for Destination GDP from the GDP Data set in the "Old" Folder

```python
countries_destination = countries_of_interest['destination_alpha_3'].unique().tolist()


missing_destination = [i for i in countries_origin if i not in gdp_c]
destination_gpd = gdp_countries[gdp_countries['Country Code'].isin(countries_destination)] 
destination_gpd = destination_gpd[gdp_columns]
destination_gpd_frame = pd.DataFrame(destination_gpd)
destination_gpd_frame.reset_index(drop=True, inplace=True)
destination_gpd_frame = destination_gpd_frame.rename(columns={'Country Code':'destination_alpha_3', '2014 [YR2014]':'destination_gdp_2014', '2015 [YR2015]':'destination_gdp_2015', '2016 [YR2016]':'destination_gdp_2016','2017 [YR2017]':'destination_gdp_2017','2018 [YR2018]':'destination_gdp_2018','2019 [YR2019]':'destination_gdp_2019'})
destination_gpd_frame
```

Following codes for the destination countries are missing in the new data
['ALA', 'GGY', 'SHN', 'AIA', 'GUF', 'GLP', 'QS', 'BES', 'XC', 'MTQ', 'QW', 'QV', 'QU'] 

Dropping the previous GDP data from the countries_of_interest dataset 

```python
countries_of_interest.drop([col for col in countries_of_interest.columns if 'gdp' in col],axis=1, inplace=True)
```

Merging Origin GDP into countries_of_interest Data

```python
GDP_merged = countries_of_interest.merge(origin_gpd_frame,on = 'origin_alpha_3', how = 'left')
GDP_merged
```

Merging Destination GDP into GDP_merged Data

```python
GDP_merged = GDP_merged.merge(destination_gpd_frame,on = 'destination_alpha_3', how = 'left')
GDP_merged
```

Removing '..' values from the data type

```python
GDP_merged.replace(to_replace = '..', value = '', inplace = True)
```

Changes the Datatype of the merged columns and dropping ''Unnamed: 0' column

```python
GDP_merged.drop(['Unnamed: 0'],axis=1, inplace=True)
GDP_merged[['destination_gdp_2014', 'destination_gdp_2015','destination_gdp_2016','destination_gdp_2017','destination_gdp_2018','destination_gdp_2019', 'origin_gdp_2014', 'origin_gdp_2015', 'origin_gdp_2016','origin_gdp_2017','origin_gdp_2018','origin_gdp_2019']] = GDP_merged[['destination_gdp_2014', 'destination_gdp_2015','destination_gdp_2016','destination_gdp_2017','destination_gdp_2018','destination_gdp_2019', 'origin_gdp_2014', 'origin_gdp_2015', 'origin_gdp_2016','origin_gdp_2017','origin_gdp_2018','origin_gdp_2019']].apply(pd.to_numeric)
```

Creating new data set file in a "New" folder

```python
GDP_merged.to_csv("New/merged_Dataset.csv", index=False)
```
